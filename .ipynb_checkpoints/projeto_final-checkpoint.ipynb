## Quais são os imóveis que a House Rocket deveria comprar e por qual preço ?


```python

# ================================================================
# Processo
# - Agrupar os imóveis por região ( zipcode ).
# - Dentro de cada região, eu vou encontrar a mediana do preço do imóvel. (farei preço/sqft_lot pra ter uma base melhor)
# - Vou sugerir os imóveis que estão abaixo do preço mediano da região e que estejam e boas condições e vista para água.
# - Exemplo:
#         - Imóvel:  
#           Cod | Região       | Preço do Imóvel | Preço na Mediana | Condições | Status
#         10330 | Vila Olímpia | R$ 450.000,00   | R$ 500.000,00    | 3         | Compra
#         10335 | Vila Olímpia | R$ 750.000,00   | R$ 500.000,00    | 3         | Não Compra
# =================================================================

#setting the columns 'price_per_sqft'
for i in range(len(data)):
    data.loc[i, 'price_per_sqft'] = data.loc[i,'price']/data.loc[i,'sqft_lot']

#price_sqft_per_zipcode is a dataframe with the median price of each zipcode, which represents the region 
price_sqft_per_zipcode = data[['price_per_sqft','zipcode']].groupby('zipcode').median().reset_index()
price_sqft_per_zipcode.columns=['zipcode','median_price_zipcode']
price_sqft_per_zipcode
```

```python
#merge the two dataframes, data and price_sqft_per_zipcode, so i got the median_price__zipcode for each id of 'data'. 
data = pd.merge(data,price_sqft_per_zipcode,on='zipcode',how='left')
data
```

```python
#setting the 'status' column that indicaes if the price of the id is above or below the median of the region and the condition is 
# higher than 3 

for i, line in data.iterrows():
    if ((line['price_per_sqft']) < (line['median_price_zipcode'])) & (line['condition']>=3):
        data.loc[i,'status'] = 'buy'
    else:
        data.loc[i,'status'] = 'not_buy'
        
data.head()

```

## Uma vez a casa comprada, qual o melhor momento para vendê-las e por qual preço?


```python

# - Com os dados já tratados e organizados.
# - Agrupar os imóveis por região ( zipcode ) e por quarter (Q1, Q2, Q3, Q4).
# - Dentro de cada região e sazonalidade, eu vou calcular a mediana do preco.
# - Condições de venda:
#     1. Se o preço da compra for maior que a mediana da região + quarter.
#     O preço da venda será igual ao preço da compra + 10%
#     2. Se o preço da compra for menor que a mediana da região + quarter.
        
#         O preço da venda será igual ao preço da compra + 30%
        
#     - Exemplo:
#         - Imóvel Cod | Região       | Quarter | Preço da Mediana | Preço de Compra | Preço de Venda      | Lucro
#           10330      | Vila Olímpia | Q1     | R$ 800.000,00    | R$ 450.000,00   | R$450.000,00 + 30%  | ??
#           10335      | Vila Olímpia | Q2   | R$ 400.000,00    | R$ 500.000,00   | R$ 500.000,00 + 10% | ??
```

```python
data.head()
```

```python
#create the column month in order to set the quarters of year (Q1,Q2,Q3,Q4) wich the house was bought
data['month'] = pd.to_datetime(data['date']).dt.strftime('%m')




data['quarter'] = data['month'].apply(lambda x: 'Q1' if 1<=int(x)<=3 else
                                               'Q2' if 4<=int(x)<=6 else
                                               'Q3' if 7<=int(x)<=9 else
                                               'Q4' if 10<=int(x)<=12 else 'NA')

data.head()
```

```python
# Calculating median of price_per_sqft, grouped by quarter and zip code 
price_quarter_zipcode = data[['price_per_sqft','quarter','zipcode']].groupby(['quarter','zipcode']).median().reset_index()
price_quarter_zipcode.columns=['quarter','zipcode','price_quarter_zipcode']
price_quarter_zipcode.sort_values('price_quarter_zipcode')
```

```python
#merging with a copy of dataset, that inludes a column of price_quarter_zipcode
data1 = data.copy()

data1 = pd.merge(data1, price_quarter_zipcode, on = ['quarter','zipcode'], how ='left')
data1
```

```python
#classsificating based on if the price_per_sqft of each input is above or below the median of price_quarter_zipcode of 
# each data

# - Condições de venda:
#     1. Se o preço da compra for maior que a mediana da região + quarter.
#     O preço da venda será igual ao preço da compra + 10%
#     2. Se o preço da compra for menor que a mediana da região + quarter.
        
#         O preço da venda será igual ao preço da compra + 30%


for i, line in data1.iterrows():
    if (line['price_per_sqft']>line['price_quarter_zipcode']):
        data1.loc[i,'above_median'] = 'yes'
        data1.loc[i,'selling_price'] = line['price']*1.1
    else:
        data1.loc[i,'above_median'] = 'no'
        data1.loc[i,'selling_price'] = line['price']*1.3
        
data1[['id', 'price_per_sqft', 'quarter','price_quarter_zipcode','above_median','selling_price' ]].head()
```

### Hipotese 4 - O crescimento do preço dos imóveis YoY ( Year over Year ) é de 10%

    

```python
#     Agrupar por ano e calcular o preço medio dos imoveis agrupados
#     Clacular crescimento medio
#     Apresentar em grafico de linha a evolucao dos preços por ano de construcao
#     Apresentar em legenda a formula para taxa de crescimento anual

data['year'] = pd.to_datetime(data['date']).dt.strftime('%Y')
data['year']

by_year = data[['year','price']].groupby('year').mean().sort_values('year').reset_index()
by_year

ax = sns.barplot (x = 'year',
                  y = 'price',
                  data = by_year)

ax.bar_label(ax.containers[0])
```

### Hipotese 5 - Imóveis com 3 banheiros tem um crescimento MoM de 15%.


```python
#     Agrupar os dados pelo numero de banheiros e depois pelo data de construcao, considerando os ultimos 5 anos
#     Calcular a diferenca MoM e apresentar em um grafico de linhas

data['yr_month'] = pd.to_datetime(data['date']).dt.strftime('%Y-%m')
by_bathrooms = data.loc[data['bathrooms']==3, ['id','yr_month']]
by_bathrooms = by_bathrooms.groupby('yr_month').count().reset_index()

total_acumulated = 0 

for i, line in by_bathrooms.iterrows():
    
    total_acumulated += line['id']
    by_bathrooms.loc[i,'total_acumulated'] = total_acumulated
    
    
final_value = by_bathrooms.loc[by_bathrooms['yr_month']== by_bathrooms['yr_month'].max(),'total_acumulated']
initial_value = by_bathrooms.loc[by_bathrooms['yr_month']== by_bathrooms['yr_month'].min(),'total_acumulated']

final_value = pd.DataFrame(final_value).reset_index()
initial_value = pd.DataFrame(initial_value).reset_index()

final = final_value.loc[0,'total_acumulated']
initial = initial_value.loc[0,'total_acumulated']

growth_rate_avg = (((final/initial)**(1/12))-1)*100

sns.lineplot (by_bathrooms, x='yr_month', y='total_acumulated');
plt.xticks ( rotation = 80 );
```
